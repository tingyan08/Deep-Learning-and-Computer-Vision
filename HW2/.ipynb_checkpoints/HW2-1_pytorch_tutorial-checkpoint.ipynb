### 1. Please do the forward pass and backpropagation with a neural network as below with ***2 iterations***, the input is 2 and the target is 1. Also, calculate the loss with ***half and the sum square error***.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from collections import OrderedDict
```

```python
X = torch.tensor([2], dtype= torch.float32)
y = torch.tensor([1], dtype= torch.float32)
```

```python
# Half of the sum square error
def loss(y, pred):
    return ((pred-y)**2).sum()/2
```

```python
# Show parameters
def show_parameters(i, X, model):
    print(f"Iters {i}")
    print("Input:")
    print(X)
    for layer_name, layers in model.named_modules():
        if isinstance(layers, nn.Linear):
            for param_name, param in layers.named_parameters():
                print(f"{layer_name} {param_name}", param)
                print(f"{layer_name} {param_name} Gradient\n", param.grad)
            print(f"{layer_name} output:")
            X = layers(X)
            print(X)
    print("===============================================")
```

```python
def initialize_weights(model):
    for name, i in model.named_modules():
        if isinstance(i, nn.Linear):
            nn.init.ones_(i.weight.data)
            nn.init.ones_(i.bias.data)
```

```python
model = nn.Sequential(OrderedDict([("Layer1", nn.Linear(1, 2)), 
                       ("Layer2", nn.Linear(2, 2)), 
                       ("Layer3", nn.Linear(2, 1))]))
initialize_weights(model)
lr = 0.01
n_iters = 2
optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0)
loss_list = []
for i in range(n_iters):
    optimizer.zero_grad()
    pred = model(X)
    l = loss(pred, y)
    loss_list.append(l.detach().numpy())
    l.backward()
    show_parameters(i, X, model)
    optimizer.step()
```

```python
import matplotlib.pyplot as plt
plt.plot(range(len(loss_list)), loss_list)
plt.show()
```
