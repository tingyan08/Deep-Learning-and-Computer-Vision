### 1. Define the model and dataset 

#### 1-1. Please follow the tutorial to create a class **ConcreteDataset**, for loading the data you need and also do the Min-Max scaling to the feature and label. In this part, please predict the **strength of concrete**. 

```python
class ConcreteDataset(Dataset):
    def __init__(self, csv_path):
        df = pd.read_csv(csv_path)
        self.X = df.iloc[:, :-1].values
        self.y = df.iloc[:, -1].values
        self.X_scaler = MinMaxScaler()
        self.y_scaler = MinMaxScaler()
        self.X = self.X_scaler.fit_transform(self.X)
        self.y = self.y.reshape(len(self.y), 1)
        self.y = self.y_scaler.fit_transform(self.y)
        
        
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        feature = torch.tensor(self.X[idx], dtype=torch.float32)
        label = torch.tensor(self.y[idx], dtype=torch.float32)
        return feature, label
    
    def get_splits(self, val_size=0.1, test_size=0.1):
        test_size = int(test_size * len(self.X))
        val_size = int(val_size * len(self.X))
        train_size = len(self.X) - test_size - val_size
        return random_split(self, [train_size, val_size, test_size])
        
```

#### 2-2. By following the tutorial, try to create a class **MLP** with three hidden layers as your network architecture. 

```python
class MLP(nn.Module):
    def __init__(self, input_ch=30, first_layers=256, second_layers=64, third_layers=8):
        super(MLP, self).__init__()
        self.hidden1 = nn.Linear(input_ch, first_layers)
        self.act1 = nn.Tanh()
        self.hidden2 = nn.Linear(first_layers, second_layers)
        self.act2 = nn.Tanh()
        self.hidden3 = nn.Linear(second_layers, third_layers)
        self.act3 = nn.Tanh()
        self.hidden4 = nn.Linear(third_layers, 1)
        self.act4 = nn.Tanh()
    
    def forward(self, X):
        X = self.hidden1(X)
        X = self.act1(X)
        X = self.hidden2(X)
        X = self.act2(X)
        X = self.hidden3(X)
        X = self.act3(X)
        X = self.hidden4(X)
        X = self.act4(X)
        return X        
```

#### 2.4 Train the MLP model with same architecture and hyperpameter below and do the validation every epoch. Choose the appropriate type of loss according to the task.
##### (Note. You should record the training/validation loss every epoch)


| | Learning rate | epochs | optimizer | momentum |
|:---:|:---:|:---:|:---:|:---:|
|***Hyperparameter***| 0.01 | 500 | SGD | 0.9 |

```python
learning_rate = 0.01
criterion = nn.MSELoss()
max_epoch = 500
optimizer = optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9)
result = {"Training Loss":[], "Validation Loss":[]}
pbar = tqdm(range(max_epoch))
for i in pbar:
    training_loss = 0
    val_loss = 0
    for inputs, label in train_dl:
        optimizer.zero_grad()
        pred = model(inputs)
        loss = criterion(pred, label)
        loss.backward()
        optimizer.step()
        training_loss+=loss.detach().numpy()

    for inputs, label in val_dl:
        with torch.no_grad():
            pred = model(inputs)
            loss = criterion(pred, label)
            val_loss+=loss.numpy()
    result["Training Loss"].append(training_loss/len(train_dl))
    result["Validation Loss"].append(val_loss/len(val_dl))    
```

#### 2.5 Please draw the plot with training/validation loss and write down the observation.

```python
plt.title("Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.plot(range(len(result[f"Training Loss"])), result[f"Training Loss"], label=f"Training Loss")
plt.plot(range(len(result[f"Validation Loss"])),  result[f"Validation Loss"], label=f"Validation Loss")
plt.legend()
plt.show()
```
