#### You will train a binary classfication model in this part. The data is **Breast Cancer Wisconsin (Diagnostic) Dataset**, you can find the details of each column at https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29

```python
# Import necessory modules
%matplotlib inline
import pandas as pd
import numpy as np
import random
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score
```

```python
# For reproduce the result
torch.manual_seed(0)
random.seed(0)
np.random.seed(0)
```

### 1. Define the model and dataset 

#### 1-1. Please follow the tutorial to create a class **BrestDataset**, for loading the data you need and also do the appropriate preprocessing from the lecture. 

```python
class BreastDataset(Dataset):
    def __init__(self, csv_path):
        df = pd.read_csv(csv_path)
        self.X = df.iloc[:, 2:-1].values
        self.y = df.iloc[:, 1].values
        self.le = LabelEncoder()
        self.scaler = MinMaxScaler()
        self.X = self.scaler.fit_transform(self.X)
        self.y = self.le.fit_transform(self.y) # 0/B; 1/M
        self.y = self.y.reshape(len(self.y), 1)
        
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        feature = torch.tensor(self.X[idx], dtype=torch.float32)
        label = torch.tensor(self.y[idx], dtype=torch.float32)
        return feature, label
    
    def get_splits(self, val_size=0.1, test_size=0.1):
        test_size = int(test_size * len(self.X))
        val_size = int(val_size * len(self.X))
        train_size = len(self.X) - test_size - val_size
        return random_split(self, [train_size, val_size, test_size])
        
```

#### 2-2. By following the tutorial, try to create a class **MLP** as your network architecture. 

```python
class MLP(nn.Module):
    def __init__(self, input_ch=30, first_layers=256, second_layers=64, third_layers=8):
        super(MLP, self).__init__()
        self.hidden1 = nn.Linear(input_ch, first_layers)
        self.act1 = nn.ReLU()
        self.hidden2 = nn.Linear(first_layers, second_layers)
        self.act2 = nn.ReLU()
        self.hidden3 = nn.Linear(second_layers, third_layers)
        self.act3 = nn.ReLU()
        self.hidden4 = nn.Linear(third_layers, 1)
        self.act4 = nn.Sigmoid()
    
    def forward(self, X):
        X = self.hidden1(X)
        X = self.act1(X)
        X = self.hidden2(X)
        X = self.act2(X)
        X = self.hidden3(X)
        X = self.act3(X)
        X = self.hidden4(X)
        X = self.act4(X)
        return X        
```

### 2. Train the model

#### 2.1 Please load the **Breast_Cancer_Wisconsin_Dataset.csv**, and split the dataset to training, validation and testing size of 0.6, 0.3, 0.1. And print the length of each dataset to check if it is right. 

```python
dataset = BreastDataset("Breast_Cancer_Wisconsin_Dataset.csv")
train_ds, val_ds, test_ds = dataset.get_splits(val_size=0.3, test_size=0.1)
```

#### 2.2 Please create the training/validation dataloder with batch size 64 and determine whether shuffle or not.

```python
train_dl = DataLoader(train_ds, batch_size=64, shuffle=True)
val_dl = DataLoader(val_ds, batch_size=64, shuffle=False)
```

#### 2.3 Create two MLP model from the table below and print the model:
##### (Note. The output layer and activation function you should determine by the task and the dataset.) 
| | First Layer | Second Layer | Third Layer | Activation between each two layers |
|:---:|:---:|:---:|:---:|:---:|
|***Model***| 32 | 16 | 8 | ReLU|

```python
model= MLP(input_ch=30, first_layers=32, second_layers=16, third_layers=8)
```

```python
print(model)
```

#### 2.4 Train the MLP model with same architecture and hyperpameter below and do the validation every epoch. Choose the appropriate type of loss according to the task.
##### (Note. You should record the training/validation loss and training/validation accuracy every epoch, the decision threshold set to 0.5)


| | Learning rate | epochs | optimizer | momentum |
|:---:|:---:|:---:|:---:|:---:|
|***Hyperparameter***| 0.01 | 300 | SGD | 0.9 |

```python
learning_rate = 0.01
criterion = nn.BCELoss()
max_epoch = 500
optimizer = optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9)
result = {"Training Loss":[], "Validation Loss":[], "Training Accuracy":[], "Validation Accuracy":[]}
pbar = tqdm(range(max_epoch))
for i in pbar:
    training_loss = 0
    val_loss = 0
    train_cf = np.zeros((2, 2))
    val_cf = np.zeros((2, 2))
    for inputs, label in train_dl:
        optimizer.zero_grad()
        pred = model(inputs)
        loss = criterion(pred, label)
        loss.backward()
        optimizer.step()
        training_loss+=loss.detach().numpy()
        cf = confusion_matrix(label.numpy(), pred.detach().numpy()>0.5)
        train_cf+=confusion_matrix(label.numpy(), pred.detach().numpy()>0.5)

    for inputs, label in val_dl:
        with torch.no_grad():
            pred = model(inputs)
            loss = criterion(pred, label)
            val_loss+=loss.numpy()
            val_cf+=confusion_matrix(label.numpy(), pred.detach().numpy()>0.5)
    result["Training Loss"].append(training_loss/len(train_dl))
    result["Validation Loss"].append(val_loss/len(val_dl))    
    result["Training Accuracy"].append((train_cf[0][0]+train_cf[1][1])/np.sum(train_cf)) 
    result["Validation Accuracy"].append((val_cf[0][0]+val_cf[1][1])/np.sum(val_cf)) 
```

#### 2.5 Please draw the plot with training/validation loss and training/validation accuracy and write down the observation.

```python
fig, ax = plt.subplots(1, 2, figsize = (16,8))
fig.subplots_adjust(hspace=0.2, wspace=0.2)
metrics = ["Loss", "Accuracy"]
for i, met in enumerate(metrics):
    ax[i].set_title(met)
    ax[i].set_xlabel("Epoch")
    ax[i].set_ylabel(met)
    ax[i].plot(range(len(result[f"Training {met}"])), result[f"Training {met}"], label=f"Training {met}")
    ax[i].plot(range(len(result[f"Validation {met}"])),  result[f"Validation {met}"], label=f"Validation {met}")
    ax[i].legend()
plt.show()
```

#### 2.6 Please calculate the confusion matrix with the test dataset. 

```python
label = []
prediction = []
for inputs, y in test_ds:
    with torch.no_grad():
        pred = model(inputs)
        label.append(y[0])
        prediction.append(int(pred[0]>0.5))
test_cf = confusion_matrix(label, prediction)
plt.matshow(test_cf, cmap=plt.cm.Blues, alpha=0.3)
for j in range(test_cf.shape[0]):
    for k in range(test_cf.shape[1]):
        plt.xlabel("Prediction")
        plt.ylabel("Actuals")
        plt.text(x=j, y=k,s=test_cf[k, j], va='center', ha='center', size='xx-large')
```

#### 2.7 Please draw the ROC curve and AUC score with the test dataset. 

```python
label = []
prediction = []
for inputs, y in test_ds:
    with torch.no_grad():
        pred = model(inputs)
        label.append(y[0])
        prediction.append(pred[0])
fper, tper, thresholds = roc_curve(label, prediction)
print(f'ROC AUC: {roc_auc_score(label, prediction):.4f}')
plt.title("Receiver Operating Characteristic Curve")
plt.plot([0, 1], [0, 1], color='black', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot(fper, tper)
plt.show()
```
